```python
import numpy as np
gfg = np.matrix('[4,1,9; 12,3,1; 4,5,6]')
print(gfg.sum())
print("row-wise ", gfg.sum(axis=1))
print("column-wise ", gfg.sum(axis=0))
```

```python
arr = np.array([10,52,62,16,16,54,453])
#(arr[::-1])

sort = np.sort(arr)
print(sort)

sort_indices = np.argsort(arr)
print(sort_indices)

smallest_4 = np.sort(arr)[:4]
print(smallest_4)

largest_5 = np.sort(arr)[-5:][::-1]
print(largest_5)
```

```python
arr = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

print(arr[arr==arr.astype(int)])
print(arr[arr!=arr.astype(int)])
```

```python
import numpy as np
x = 77+75
sales = np.array([x,x+50,x+100,x+150,x+200])
print(sales)

tax_rate = ((x%5)+5)/100
tax = sales + (sales * tax_rate)
print(tax)

dis = np.where(sales<x+100, sales*0.95, sales*0.90)
print("after discount", dis)

weekly_sales = np.tile(sales, (3,1))
print(weekly_sales)

inc_rates = np.array([1.00,1.02,1.04]).reshape(3,1)
adjusted_sales = weekly_sales * inc_rates
print(adjusted_sales)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)


plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x, y1, color='blue')
plt.title("y = xÂ²")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(x, y2, color='green')
plt.title("y = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(x, y3, color='red')
plt.title("y = e^x")
plt.xlabel("x")
plt.ylabel("y")
plt.yscale('log')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(x, y4, color='purple')
plt.title("y = log(|x| + 1)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.tight_layout()
plt.show()
```
