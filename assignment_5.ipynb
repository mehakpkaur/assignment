```python
import numpy as np
gfg = np.matrix('4,1,9;12,3,1;4,5,6')
print("sum of all elements", np.sum(gfg))
print("sum of all elements row-wise", np.sum(gfg, axis =1))
print("sum of all elements column-wise", np.sum(gfg, axis =0))
```

```python
import numpy as np
array = np.array([10,52,62,16,16,54,453])
print("sorted array", np.sort(array))
print("indices of sorted array",np.argsort(array))
print("4 smallest elements", np.sort(array)[:4])
print("5 largest elements", np.sort(array)[-5:])
```

```python
import numpy as np
array = np.array([1.0,1.2,2.2,2.0,3.0,2.0])
intresult = array[array == array.astype(int)]
print("integer elements only", intresult)
floatresult = array[array != array.astype(int)]
print("float elements only", floatresult)
```

```python
import numpy as np
#(a)
X= ord("M") + ord("K")
print("ascii sum : ", X)
sales = np.array([X, X+50, X+100, X+150, X+200])
print("sales dataset", sales)

#(b)
taxrate = ((X%5)+5)/100
print("taxrate", taxrate)

#(c)
discounted_sales = []
for s in sales :
  if s<(X+100):
    discounted_sales.append(s * (1-0.05))
  else:
    discounted_sales.append(s*(1-0.10))
discounted_sales = np.array(discounted_sales)
print("discounted sales",discounted_sales )

#(d)
weekly_sales = np.tile(sales,(3,1))
growth_factors = np.array([1.00,1.02,1.04]).reshape(3,1)
adjusted_sales = weekly_sales * growth_factors
print("adjusted sales", adjusted_sales)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10, num = 100)

y1 = x*x
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x)+1)

plt.figure(figsize =(10,6))

plt.plot(x, y1, label = 'Y = x^2', color = 'blue')
plt.plot(x, y2, label = 'Y = sin(x)', color = 'green')
plt.plot(x, y3, label = 'Y = exp(x)', color = 'red')
plt.plot(x, y4, label = 'Y = log(|x|=1)', color = 'purple')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Plots of different mathematical functions")
plt.grid(True)
plt.legend()
plt.show()
```
