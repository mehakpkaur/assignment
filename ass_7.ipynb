```python
import pandas as pd
data = {
    "Employee_Id": [101,102,103,104,105],
    "Name":["Alice", "Bob", "Charlie", "Diana", "Edward"],
    "Department": ["HR", "IT", "IT", "Marketing", "Sales"],
    "Age": [29,34,41,28,38],
    "Salary": [50000, 700000, 65000, 55000, 60000],
    "Years_of_Experience": [4,8,10,3,12],
    "Joining_Date": ["2020-03-15", "2017-07-19", "2013-06-01", "2021-02-10", "2010-11-25"],
    "Gender": ["Female", "Male", "Male","Female", "Male"],
    "Bonus": [5000, 7000, 6000, 4500, 5000],
    "Rating": [4.5, 4.0, 3.8, 4.7, 3.5]
 }
df = pd.DataFrame(data)
print(df)

#(a)
print("Shape of DataFrame", df.shape)
#(b)
print("Summary ", df.info())
#(c)
print("Descriptive Statistics", df.describe())
#(d)
print("First five rows", df.head())
print("Last three rows", df.tail(3))
#(e)
print("Average salary", df["Salary"].mean())
print("Total Bonus Paid", df["Bonus"].sum())
print("Youngest employee age", df["Age"].min())
print("Highest Performance Rating", df["Rating"].max())
#(f)
df_sorted = df.sort_values(by = "Salary", ascending = False)
print("Descending order of salary:-")
print(df_sorted)
#(g)
def categorize_rating(rating):
    if rating >= 4.5:
        return "Excellent"
    elif 4.0 <= rating < 4.5:
        return "Good"
    else:
        return "Average"

df["Rating_Category"] = df["Rating"].apply(categorize_rating)
print("\nDataFrame with Rating Categories:")
print(df)
#(h)
print("Missing Values")
print(df.isnull())
#(i)
df.rename(columns = {"Employee_Id": "Id"}, inplace = True)
print("Renamed Columns")
print(df.columns)
#(j)
exp_mor_than_5 = df[df["Years_of_Experience"] > 5]
it_department = df[df["Department"] == "IT"]
print("Employees with experience more than 5 years:-")
print(exp_mor_than_5)
print("IT Department Employees:-")
print(it_department)
#(k)
df["Tax"] = df["Salary"] * 0.10
print("DataFrame with Tax Column :")
print(df)
#(l)
df.to_csv("modified_employees.csv", index=False)
print("\nModified DataFrame saved as 'modified_employees.csv'")
```
